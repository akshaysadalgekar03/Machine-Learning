### BATCH GRADIENT DESCENT

**Model trained using stochastic gradient descent with a batch size of 32 and mean squared error as your loss function.**

```python
class linear_regression:
    def __init__(self):
        self.w = 0
        self.b =0
        

    def batch_gradient_descent(self, X_new, Y_new):
        self.w = 0
        self.b = 0
        iterations = 100 
        X = X_new
        Y = Y_new
        plt_mse = []
        number = []
        for _ in range(iterations):
            for start in range(0, len(X)+1, 32):    
                stop = start + 32
                xb, yb = X[start:stop], Y[start:stop]
                Yp = self.w * xb + self.b  
                c_loss = np.sum((yb-Yp)**2)/len(yb)
                plt_mse.append(c_loss)
                self.w = self.w - 0.001 * ((-2/len(xb)) * sum(xb * (yb - Yp)))  
                self.b = self.b - 0.001 * ((-2/len(xb)) * sum(yb - Yp))  
                
                current_batch = stop - start
                number.append(_)
                print("batch_size :", len(xb))
                print("Step number: ",_+1," Loss is :",plt_mse[_])
        
        plt.figure(figsize=(14,12))
        plt.plot(number, plt_mse)
        plt.scatter(number, plt_mse, marker="*", c= '#bcbd22', cmap='Greens', edgecolors="Orange")
        plt.xlabel("Step number", fontsize = 12)
        plt.ylabel("Loss", fontsize = 12)
        plt.show()
        plt.show()

    
    def predict(self, x, y):
        yp = x * self.w + self.b
        ya = y 
        current_cost = np.sum((ya-yp)**2)/len(ya)
        print("cost for testing set: ",current_cost)

 
```

### REGULARIZATION:

**To observe the effects of regularization, picked one of trained models and inspect the weights. 
Trained an identical model again, except this time with L2 regularization to the loss**


```python
class linear_regression_regularized:
    def __init__(self):
        self.w = 0
        self.b =0
        

    def batch_gradient_descent_l2(self, X_new, Y_new):
        self.w = 0
        self.b = 0
        iterations = 100  
        X = X_new
        Y = Y_new
        plt_mse = []
        number = []
        for _ in range(iterations):
            for start in range(0, len(X)+1, 32):    
                stop = start + 32
                xb, yb = X[start:stop], Y[start:stop]
                Yp = self.w * xb + self.b  # The current predicted value of Y
                c_loss = np.sum((yb-Yp)**2)/len(yb) 
                plt_mse.append(c_loss)

                self.w = self.w - 0.0001 * ((-2/len(xb)) * sum(xb * (yb - Yp)) + (2) * (self.w**2)) 
                self.b = self.b - 0.0001 * ((-2/len(xb)) * sum(yb - Yp))  
                
                current_batch = stop - start
                number.append(_)
                print("batch_size :", len(xb))
                print("iteration: ",_+1," Loss: ",plt_mse[_],)
        
        # plot mse
        plt.figure(figsize=(14,12))
        plt.plot(number, plt_mse)
        plt.scatter(number, plt_mse, marker="*", c= '#bcbd22', cmap='Greens', edgecolors="Orange")
        plt.xlabel("Step number", fontsize = 12)
        plt.ylabel("Loss", fontsize = 12)
        plt.show()
        plt.show()
 


    def predict_l2(self, x, y):
        yp = x * self.w + self.b
        ya = y 
        current_cost = np.sum((ya-yp)**2)/len(ya)
        print("cost for testing set: ",current_cost)

 
```

```python
linear_using_l2 = linear_regression_regularized()
linear_using_l2.batch_gradient_descent_l2(X_train[:, 0], X_train[:, 1])
linear_using_l2.predict_l2(X_test[:, 0], X_test[:, 1])
```
